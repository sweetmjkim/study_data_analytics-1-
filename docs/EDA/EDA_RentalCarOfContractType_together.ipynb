```python
import pandas as pd
df_RCOCT = pd.read_csv('RentalCarOfContractType.csv')
df_RCOCT.info()
df_RCOCT.dropna(inplace=True)
```

### 기술 데이터 분석(DDA)

```python
# 계약 유형 분포(범주형) 확인
df_RCOCT['type_of_contract'].value_counts()
# 대부분 고객은 렌탈 계약 유형을 선택
# 인사이트 : 렌탈 계약 유형에 대한 고객 선호도가 높음 반영, 마케팅 및 프로모션 전략 고려 할 필요 있음.
```

```python
# 고객 연령 분포 확인
df_RCOCT['age'].describe()
# 소비 고객 연령 대부분 중년층 (46세 전후)에 모여 있음
# 인사이트 : 한정된 연령대 소비자가 있으므로, 소비자 층을 넓히기 위한  전략 수립 필요
```

```python
# 마일리지 분포 확인
df_RCOCT['Mileage'].describe()
# 평균(중앙값) 마일리지는 약 1946.00이며, 대부분이 Q1과 Q3사이에 있고, 최소 872.0, 최대 3803.0 사이 분포
# 인사이트 : 고객 마일리지 사용 패턴을 분석해 로열티 프로그램보다 많이 개발 활용
```
