### 수치형 데이터 타입 판단
- count : Null 없는 series 개수
- mean : 평균
-  std : standard deviation
-  50% : median 중앙값
-  min, max

```python
df_titanic.describe() # 수치형만 표현

# PassengerID : 통계로 알아낼 수 있는 값이 없음
# Pclass : min 과 max와 구성값의 갯수를 보았을 때 범주형 타입으로 사용 필요
# Age : 흩어져서 분석 결과가 모호한 경우, 일정 범위로 묶어 범주형으로 만듦
# Fare : 분석 시에는 수치형, 시각화시에는 범주형
```

```python
# min 과 max와 구성값의 갯수를 보았을 때 범주형 타입으로 사용 필요
df_titanic['Pclass'].unique()
```

```python
# 흩어져서 분석 결과가 모호한 경우, 일정 범위로 묶어 범주형으로 만듦
df_titanic['Age'].unique()
```

### 범주형 데이터 타입 판단
- unique : 고유한 값 개수
- top : 고유한 값에서 최상값(문자 : 오름차순 처음, 숫자 : 오름차순 끝)
- freq : 가장 많은 반복횟수

```python
df_titanic.describe(include=object) # 범주형 파악
# Name 분석에 맞지 않음. 성씨를 범주형 변환 시 사용 가능
```
