```python
import pandas as pd
df_RCOCT = pd.read_csv("/content/RentalCarOfContractType.csv")
df_RCOCT[:2] 
```

## X (범주형)/ Y (범주형) : 주로 독립성 증명
- 귀무가설 : 관계성이 없다
- 대립가설 : 관계성이 있다 -> 주로 연구원이나 분석가가 증명하고 싶은 것
  

### 예제 가설 : type_of_contract, cancellation


#### 분석 내용 : 계약 유형에 따라 해약여부가 달라지는가?
  + 귀무가설 : 계약 유형에 따라 해약여부가 달라지지 않는다. (p-value > 0.05)
  + 대립가설 : 계약 유형에 따라 해약여부가 달라진다.(p-value <= 0.05)

```python
df_RCOCT_category_pivot = pd.pivot_table(data = df_RCOCT[['type_of_contract','cancellation']], index='type_of_contract',columns = ['cancellation'], aggfunc=len)
df_RCOCT_category_pivot
```

```python
import scipy
import scipy.stats as stats
```

```python
# correction = True : 데이터 수집 불균형을 완화 하기 위한 param --> 유의사항 : 결측치 없어야 한다.
stats.chi2_contingency(observed=df_RCOCT_category_pivot, correction = True)
# expected_freq : 기대빈도수, dof : 자유도, statistic : 통계 수치, p(probability) : 확률
```

#### 분석 결과( 기준 : p-value 0.05)
- 통계 측면 결론 : 대립 가설이 성립, 귀무 가설 기각
- 사실 결론 : 계약 유형에 따라 해약여부가 달라진다.
