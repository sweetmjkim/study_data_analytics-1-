## load dataset

```python
import pandas as pd
df_RCOCT = pd.read_csv("/content/RentalCarOfContractType.csv")
df_RCOCT[:2]
```

## 범주형 / 연속형 : 범주형에 따른 관계 확인

### 연속형 항목이 정규 분포이며 ,등분산일 때

### 예제 가설 : channel(항목 2), Mileage[:3000](정규분포)
- 분석 내용 : Channel 중 서비스 방문과 홈쇼에 따른 마일리지 평균 차이가 있는가
  + 귀무 가설 : 평균 차이가 없다.
  + 대립 가설 : 평균 차이가 있다.

#### T test Test(등분산)
- 연속형이 정규분포이면서 범주형에 따른 항목이 등분산, 항목 갯수가 2개

```python
# 필요 항목 확인
channel_list = df_RCOCT['channel'].value_counts().iloc[:2].index.to_list()
channel_list
```

```python
series_list = list()
for channel in channel_list:
  condition = f"channel == '{channel}'"
  series_Mileage = df_RCOCT.query(condition)['Mileage'].dropna() # condition에 따른 series 작성
  series_list.append(series_Mileage)
series_list[1].iloc[:5]
```

```python
# 등분산 검증
import scipy.stats as stats
stats.levene(series_list[0],series_list[1]) # 등분산이다.
```

```python
# T test
stats.ttest_ind(*series_list,equal_var = True) # 등분산일 때 사용
```

#### ANOVA Test
- 연속형이 정규분포이면서 범주형에 따른 항목이 등분산, 항목 갯수가 3개 이상

```python
# 필요 항목 확인
channel_list = df_RCOCT['channel'].value_counts().iloc[:3].index.to_list()
channel_list
```

```python
series_list = list()
for channel in channel_list:
  condition = f"channel == '{channel}'"
  series_Mileage = df_RCOCT.query(condition)['Mileage'].dropna() # condition에 따른 series 작성
  series_list.append(series_Mileage)
series_list[1].iloc[:5]
```

```python
stats.f_oneway(*series_list)
```

#### T-test Tes(이분산)
- 연속형이 정규분포이면서 범주형에 따른 항목이 이분산, 항목 갯수가 2개

```python
# T test
stats.ttest_ind(series_list[0],series_list[1],equal_var = False) # 이분산일 때 사용
```
