```python
import pandas as pd
df_RCOCT = pd.read_csv("/content/RentalCarOfContractType.csv")
df_RCOCT[:2]
```

## 연속형과 연속형

### 상관 계수(correlation coefficient) 확인
- -1 <= x <=1 : 0.5 or -0.5 기준
- 0에 가까울 수록 관계 없음, 1(정비례) or -1(반비례) 쪽으로 가까울 수록 관계 있음

#### 예제 가설 : age, Mileage

### 분석 내용 : 나이에 따라 마일리지 변화가 있는가
- 귀무 가설 : 나이에 따라 마일리지 변화가 없다.
- 대립 가설 : 나이에 따라 마일리지 변화가 있다.

```python
# 결측치 확인
df_RCOCT[['age','Mileage']].isnull().sum()
```

```python
df_RCOCT_age_mileage = df_RCOCT[['age','Mileage']].dropna().copy()
df_RCOCT_age_mileage.isnull().sum()
```

```python
# 연관 계수 확인
df_RCOCT_age_mileage.corr()
```

### 정규분포 여부

#### 변수 둘 다 정규분포 : pearsonr

```python
import scipy.stats as stats
stats.pearsonr(df_RCOCT_age_mileage['age'],df_RCOCT_age_mileage['Mileage'])
```

### 분석 결과
- 통계 결론 : 대립 가설 성립, 상관계수가 정비례 관계
- 사실 결론 : 나이에 따라 마일리지 변화가 있다.

##### 변수 둘 중 하나가 비정규분포 : spearmanr test

```python
stats.spearmanr(df_RCOCT_age_mileage[:400]['age'],df_RCOCT_age_mileage[:400]['Mileage'])
```
