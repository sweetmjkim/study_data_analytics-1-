### 과일 유형(X)과 고객 세그먼트(Y) 사이에 연관성이 있습니까?
- 분석 방법: 카이제곱 독립성 검정을 사용하여 서로 다른 고객 세그먼트에 걸쳐 과일 유형의 판매 분포가 독립적인지 평가합니다.



### 가설 : 물품소분류, 연령대


#### 분석 내용 : 계약 유형에 따라 해약여부가 달라지는가?
  + 귀무가설 : 과일 유형에 따라 고객 연령대가 달라지지 않는다. (p-value > 0.05)
  + 대립가설 : 과일 유형에 따라 고객 연령대가 달라진다.(p-value <= 0.05)

```python
import pandas as pd
df_fruit = pd.read_csv("/content/fruitstore_saleshistory.csv")
condition_fruit = "물품중분류 == '과일'"
df_fruit=df_fruit.query(condition_fruit)
```

```python
df_fruit.isnull().sum()
```

```python
df_fruit_category_pivot = pd.pivot_table(data = df_fruit[['연령대','물품소분류']], index='물품소분류',columns = ['연령대'], aggfunc=len)
df_fruit_category_pivot.fillna(0,inplace=True)
df_fruit_category_pivot
```

```python
import scipy
import scipy.stats as stats
```

```python
# correction = True : 데이터 수집 불균형을 완화 하기 위한 param --> 유의사항 : 결측치 없어야 한다.
stats.chi2_contingency(observed=df_fruit_category_pivot, correction = True)
# expected_freq : 기대빈도수, dof : 자유도, statistic : 통계 수치, p(probability) : 확률
```

```python
round(1.3037916341809125e-37,5)
```

#### 분석 결과( 기준 : p-value 0.05)
- 통계 측면 결론 : 대립 가설이 성립, 귀무 가설 기각
- 사실 결론 : 계약 유형에 따라 해약여부가 달라진다.
