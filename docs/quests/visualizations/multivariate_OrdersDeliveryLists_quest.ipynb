```python
import pandas as pd
df_shopping = pd.read_csv('/content/ShoppingMallDeliveryWithDate.csv')
def convert_date(x):
    date, ampm, time = x.split()
    return pd.to_datetime(f"{date}", format="%Y-%m-%d")
df_shopping['배송시작일'] = df_shopping['배송시작일'].apply(convert_date)
df_shopping['배송완료일'] = df_shopping['배송완료일'].apply(convert_date)
df_shopping
```

### 1. 상품구매금액이 80,000원 이상인 주문들을 찾으세요.

```python
condition_first = '상품구매금액 >= 80000'
df_shopping.query(condition_first).sort_values(by='상품구매금액', ascending=True)
```

### 2. 모바일웹을 통해 주문된 상품 중, 상품구매금액이 평균 이상인 주문들을 찾으세요.

```python
import numpy as np
condition_second_route = '주문경로 == "모바일웹"'
rout_average = pd.pivot_table(data=df_shopping[['주문경로','상품구매금액']],index=['주문경로'],values=['상품구매금액'],aggfunc = np.mean)
num_average = rout_average.loc["모바일웹","상품구매금액"]
condition_second_average = f"상품구매금액 >= {num_average}"
conditions = f'{condition_second_route} and {condition_second_average}'
df_shopping.query(conditions).sort_values(by='상품구매금액', ascending=True)
```

### 3. 2019년 1월 7일 이전에 배송이 시작된 주문들을 찾으세요.

```python
df_shopping['배송시작일_int'] = df_shopping['배송시작일'].dt.year*10000+df_shopping['배송시작일'].dt.month*100+df_shopping['배송시작일'].dt.day
condition_third = '배송시작일_int < 20190107'
df_shopping.query(condition_third).sort_values(by='배송시작일', ascending=False)
```

### 4. 배송 시작일과 배송 완료일이 3일 이상 차이 나는 주문들을 찾으세요.

```python
df_shopping['배송기간'] = df_shopping['배송완료일'] - df_shopping['배송시작일']
df_shopping['배송기간'] = df_shopping['배송기간'].dt.days
df_shopping['배송기간']
condition_forth = '배송기간 >= 3'
df_shopping.query(condition_forth).sort_values(by='배송기간', ascending=True)
```
